```python
import pandas as pd
import matplotlib.pyplot as plt

# Estilo do gráfico (preview deles na documentação do Matplotlib)

# plt.style.use("dark_background")
# plt.style.use("ggplot")
# plt.style.use("bmh")
# print(plt.style.available)

# plt.style.use("ggplot")
# plt.style.use("seaborn-v0_8")
# plt.style.use("dark_background")
# plt.style.use("bmh")
# plt.style.use("fivethirtyeight")

URL = (
    "https://raw.githubusercontent.com/"
    "cfneves/turma-visualizacao-de-dados/"
    "master/aulas/semana_04/bases/base_rh.xlsx"
)

df = pd.read_excel(URL)

df["Data_Admissao"] = pd.to_datetime(
    df["Data_Admissao"],
    format="%d/%m/%Y",
    errors="coerce"
)

# Tamanho da figura
plt.figure(figsize=(10,5))

# Histograma
plt.hist(
    df["Salario"],
    bins=10,
    color="steelblue",
    edgecolor="white"
)

# Títulos
plt.title("Distribuição de Salários", fontsize=14)
plt.xlabel("Salário (R$)")
plt.ylabel("Quantidade de Funcionários")

plt.show()
```
