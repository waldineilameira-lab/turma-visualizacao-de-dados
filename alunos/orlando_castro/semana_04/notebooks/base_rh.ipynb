````python
# Datatime
# strptime() → converte string para datetime
# strftime() → converte datetime para string

from datetime import datetime

# strptime() → converte string para datetime
data_texto = "15/08/2024"
data_obj = datetime.strptime(data_texto, "%d/%m/%Y")  # '%d/%m/%Y' indica o formato da data (dia/mês/ano)

print(f'String original : "{data_texto}"')
print(f'Objeto datetime : {data_obj})')
print(f'Tipo            : {type(data_obj)}')
print()

# strftime() → converte datetime para string
formatado = data_obj.strftime('%d de %B de %Y')  # Exemplo: '15 de agosto de 2020'
print(f'String formatada: "{formatado}"')

# Extraindo partes da data
print(f'Dia             : {data_obj.day}')
print(f'Mês             : {data_obj.month}') 
print(f'Ano             : {data_obj.year}')
````

````python
# Timedelta → representa uma duração (diferença entre datas)
from datetime import timedelta
hoje = datetime.today()  # Data e hora atual

diferenca = hoje - data_obj # Subtração entre datetimes resulta em um timedelta

print(f'Data de Adminssão: {data_texto}')
print(f'Data atual       : {hoje.strftime("%d/%m/%Y")}')
print(f'Diferença em dias: {diferenca.days} dias')
print(f'Diferença em anos: {diferenca.days / 365.25:.1f} anos')
````

````python
# Convertendo uma coluna inteira para datetime
# Antes de converter posicionar na primeira linha pelo indice (iloc = integer location) 

print(f'Tipo ANTES da conversão: {df["Data_Admissao"].dtype}')
print(f'Valor ANTES da conversão: {df["Data_Admissao"].iloc[0]}')

# Fazer uma cópia do DataFrame para evitar modificar o original
df_limpo = df.copy()

# Converter a coluna "Data_Admissao" para datetime
df_limpo['Data_Admissao'] = pd.to_datetime(
    df_limpo['Data_Admissao'], 
    format="%d/%m/%Y",          # formato da string de entrada
    errors='coerce'             # datas inválidas viram NaT (Not a Time) em vez de gerar erro    
)

print(f'Tipo DEPOIS da conversão: {df_limpo["Data_Admissao"].dtype}')
print(f'Valor DEPOIS da conversão: {df_limpo["Data_Admissao"].iloc[0]}')
print(f'Número de datas inválidas (NaT): {df_limpo["Data_Admissao"].isna().sum()}')
````

````python
# Extraindo informações da data com o acesso ao atributo .dt

# Criar colunas derivadas de data
df_limpo['Ano_Admissao'] = df_limpo['Data_Admissao'].dt.year
df_limpo['Mes_Admissao'] = df_limpo['Data_Admissao'].dt.month
df_limpo['Trimestre'] = df_limpo['Data_Admissao'].dt.quarter

hoje_ts = pd.Timestamp.today()  # Data e hora atual como Timestamp do pandas

df_limpo['Tempo_Casa_Anos'] = (
    (hoje_ts - df_limpo['Data_Admissao']).dt.days / 365.25).round(1)

#Exibe as 5 primeiras linhas do DataFrame com as novas colunas de data
cols = ['Nome', 'Data_Admissao', 'Ano_Admissao', 'Mes_Admissao', 'Trimestre', 'Tempo_Casa_Anos' ]
df_limpo[cols].head(6)
````

````python
# Criando sua primeira função

def classificar_tempo(anos):
    # pd.isna() verifica se o valor é nulo (NaN ou NaT)
    if pd.isna(anos):
        return 'Sem data'
    elif anos < 1:
        return 'Menos de 1 ano'
    elif anos < 3:
        return '1 a 3 anos'
    elif anos < 5:
        return '3 a 5 anos'
    else:
        return 'Mais de 5 anos'


# Testar a função com valores individuais antes de aplicar no DataFrame
print('Testando a função:')
print(f'  0.5 anos → "{classificar_tempo(0.5)}"')
print(f'  2.1 anos → "{classificar_tempo(2.1)}"')
print(f'  4.8 anos → "{classificar_tempo(4.8)}"')
print(f'  9.0 anos → "{classificar_tempo(9.0)}"')
````

````python
# .apply(função) → aplica a função em cada linha da coluna
# Aplica a função classificar_tempo na coluna Tempo_Casa_Anos e cria uma nova coluna Faixa_Tempo_Casa com o resultado

df_limpo['Faixa_Tempo_Casa'] = df_limpo['Tempo_Casa_Anos'].apply(classificar_tempo)

print('Distribuição por faixa de tempo de casa:')
print(df_limpo['Faixa_Tempo_Casa'].value_counts())
````

````python
# Filtrando por periodo de datas

hoje = pd.Timestamp.today()
um_ano_atras = hoje - pd.DateOffset(years=1)  # Data de um ano atrás
cinco_anos_atras = hoje - pd.DateOffset(years=5)  # Data de cinco anos atrás

#Filtrar por periodo
novos = df_limpo[df_limpo['Data_Admissao'] >= um_ano_atras]
veteranos = df_limpo[df_limpo['Data_Admissao'] < cinco_anos_atras]

print(f'Funcionários admitidos no último ano: {len(novos)}')
print(f'Funcionários admitidos há mais de 5 anos: {len(veteranos)}')

print()

# Comparar salário médio entre os grupos
print('Salário médio:')
print(f'  Novos (último ano)   : R$ {novos["Salario"].mean():.2f}')
print(f'  Veteranos (> 5 anos) : R$ {veteranos["Salario"].mean():.2f}')
````

````python
# Agrupamentos com groupby() → permite agrupar os dados por uma ou mais colunas e calcular estatísticas para cada grupo

# Salário médio por departamento

print('Salário médio por departamento:')
print(
    df_limpo.groupby('Departamento')['Salario']
    .mean()
    .round(2)
    .sort_values(ascending=False)
)
````

````python
# Tempo médio de casa por departamento

media_depto = (
    df_limpo
    .groupby('Departamento')['Tempo_Casa_Anos']
    .mean()
    .round(1)
    .sort_values(ascending=False)
    # rename() → renomeia a série resultante para um nome mais descritivo
    .rename('Media_Anos')  
    .reset_index()         
)

print('Tempo médio de casa por departamento:')
print(media_depto.to_string(index=False))

# .idxmax() → retorna o NOME do índice com maior valor
maior = media_depto.set_index('Departamento')['Media_Anos'].idxmax()
menor = media_depto.set_index('Departamento')['Media_Anos'].idxmin()
print(f'\n→ Maior retenção: {maior}')
print(f'→ Menor retenção: {menor}')
````

````python
# Admissões por ano — histórico completo
# .size() → conta o número de linhas em cada grupo
# .sort_index() → ordena pelo índice (neste caso, o ano)

por_ano = (
    df_limpo
    .groupby('Ano_Admissao')
    .size()
    .rename('Admissoes')
    .reset_index()
    .dropna()  # remove linhas com Ano_Admissao nulo
)

print('Admissões por ano:')
for _, row in por_ano.iterrows():
    # .iterrows() → itera linha por linha, retornando (índice, Series)
    # _ → descartamos o índice (convenção: _ = 'não preciso desse valor')
    barra = '█' * int(row['Admissoes'] / 5)
    print(f'  {int(row["Ano_Admissao"])}  {row["Admissoes"]:>4}  {barra}')
````

````python
# Salvar o DataFrame enriquecido
# encoding='utf-8-sig' → UTF-8 com BOM — necessário para o Excel no Windows
#                         abrir o arquivo sem distorcer ã, é, ç

df_limpo.to_csv(
    'base_rh_dia01.csv',
    index=False,         # não salva o índice numérico como coluna extra
    sep=';',             # separador ponto-e-vírgula
    encoding='utf-8-sig'
)

print('✓ base_rh_dia01.csv salvo!')
print(f'  Colunas finais: {df_limpo.columns.tolist()}')
print(f'  Linhas        : {len(df_limpo)}')
````
