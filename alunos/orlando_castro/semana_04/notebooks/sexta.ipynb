```python
import pandas as pd

#--- LEITURA DO ARQUIVO CSV ---
df = pd.read_csv("C:/Users/Dell/Documents/GitHub/turma-visualizacao-de-dados/alunos/orlando_castro/semana_04/bases/base_rh.csv", sep = ";", encoding="cp1252", decimal=",")

#--- EXPLORAÇÃO DOS DADOS ---
df.shape        #(1000, 10) - 1000 linhas e 10 colunas
df.head(5)      #exibe as 5 primeiras linhas do DataFrame
df.tail(5)      #exibe as 5 últimas linhas do DataFrame
df.info()       #exibe informações sobre o DataFrame, como o número de linhas, colunas, tipos de dados e valores nulos
df.describe()   #exibe estatísticas descritivas para as colunas numéricas do DataFrame
df.nunique()    #exibe o número de valores únicos para cada coluna

df.columns.value_counts()   #exibe a contagem de valores únicos para a coluna especificada
df.isnull().sum()           #exibe a contagem de valores nulos para cada coluna

#--- SELEÇÃO E FILTROS ---
df['Nome']      #seleciona a coluna "Nome" do DataFrame
df[['Nome', 'Idade']]   #seleciona as colunas "Nome" e "Idade" do DataFrame
df[df['Status'] == 'Ativo']    #filtra as linhas onde o status é "Ativo"
df[(df['Status'] == 'Ativo') & (df['Idade'] > 30)]   #filtra as linhas onde o status é "Ativo" e a idade é maior que 30 

#--- DATAS ---
df['Data_Admissao'] = pd.to_datetime(df['Data_Admissao'], format='%d/%m/%Y', errors='coerce')   #converte a coluna "Data_Admissao" para o formato datetime
df['Ano'] = df['Data_Admissao'].dt.year   #extrai o ano da coluna "Data_Admissao" e cria uma nova coluna "Ano"
df['Mes'] = df['Data_Admissao'].dt.month  #extrai o mês da coluna "Data_Admissao" e cria uma nova coluna "Mes"
df['Tri'] = df['Data_Admissao'].dt.quarter  #extrai o trimestre da coluna "Data_Admissao" e cria uma nova coluna "Tri"
df['Tempo'] = ((pd.Timestamp.today() - df['Data_Admissao']).dt.days/365.25).round(1)  #calcula o tempo em dias desde a data de admissão até hoje e cria uma nova coluna "Tempo"

novos = df[df["Data_Admissao"]>= pd.Timestamp.today()- pd.DateOffset(years=1)]   #filtra as linhas onde a data de admissão é maior ou igual a um ano atrás a partir de hoje

#---GROPUPBY E AGREGAÇÃO ---
df.groupby('Departamento')['Salario'].mean().round(2)   #agrupa os dados por departamento e calcula a média do salário
df.groupby('Departamento')['Salario'].size().sort_values(ascending=False)   #agrupa os dados por departamento e conta o número de ocorrências, ordenando em ordem decrescente

#--- EXPORTAÇÃO DO DATAFRAME ---

df.to_csv("saida.csv", index=False, sep=";", encoding="cp1252", decimal=",")   #exporta o DataFrame para um arquivo CSV, sem incluir o índice, usando ";" como separador, "cp1252" como codificação e "," como separador decimal    
df.to_json("saida.json", orient="records", force_ascii=False, indent=4)   #exporta o DataFrame para um arquivo JSON, orientado por registros, sem forçar a codificação ASCII e com indentação de 4 espaços

with pd.ExcelWriter("saida.xlsx", engine="openpyxl") as writer:   #exporta o DataFrame para um arquivo Excel usando o mecanismo "openpyxl"
    df.to_excel(writer, sheet_name="Dados", index=False)   #escreve o DataFrame na planilha "Dados" do arquivo Excel, sem incluir o índice  
```
