```python
#importar as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

#indicar o caminho do arquivo CSV
URL = "https://github.com/cfneves/turma-visualizacao-de-dados/blob/master/alunos/maria_helena/Semana04/base_rh.xlsx?raw=true"

df = pd.read_excel(URL)

#formatar data para datatime:
df["Data_Admissao"] = pd.to_datetime(
    df["Data_Admissao"],
    format="%d/%m/%Y",
    errors="coerce"
)


```

```python
#INFORMAÇÕES INICIAIS
#mostrar as 5 primeiras linhas do dataframe 
print(df.head())
```

```python
#INFORMAÇÕES INICIAIS
#mostrar tipos de dados e a existencia de valores nulos:
print(df.info())
```

```python
#INFORMAÇÕES INICIAIS
#mostrar os nomes das colunas:
print(df.columns)
```

```python
#INFORMAÇÕES INICIAIS
#mostrar estatisticas numericas:
print(df.describe())
```

```python
# filtrar uma coluna (Serie):
salarios = df["Salario"]
print(f"Tipo  : {type(salarios)}")
print(f"shape : {salarios.shape}")   # (1000,) — 1 dimensão
print(f"dtype : {salarios.dtype}")
print(salarios.head(4))

print("\n" + "─"*38)
```

```python
#filtrar duas colunas:

sub = df[["Nome", "Salario"]]
print(f"\nTipo  : {type(sub)}")
print(f"shape : {sub.shape}")         # (1000, 2) — 2 dimensões
print(sub.head(3))

print("\n" + "─"*38)
```

```python
#mostrar dados estatisticos de uma serie (uma coluna), no caso, de salarios:
salarios = df["Salario"]

#O `:,.2f` dentro do f-string formata o número com separador de milhar (`,`) e duas casas decimais (`.2f`). Exemplo: `12500.50` vira `12,500.50`.
print(f"\nMédia   : R$ {salarios.mean():,.2f}")
print(f"Mediana : R$ {salarios.median():,.2f}")
print(f"Mínimo  : R$ {salarios.min():,.2f}")
print(f"Máximo  : R$ {salarios.max():,.2f}")
```

```python
#histograma de salarios:
# bins=10 divide o intervalo em 10 faixas iguais
plt.hist(df["Salario"], bins=10, color="steelblue")     #cria o grafico
plt.title("Distribuição de Salários")                   #adiciona um título
plt.xlabel("Salário (R$)")                              #titulo eixo "x"
plt.ylabel("Quantidade de Funcionários")                #titulo eixo "y"    
plt.show()                                              #exibe o grafico
```

```python
#histograma salarios de funcionarios ativos (dois filtros):
salarios_ativos = df[df["Status"] == "Ativo"]["Salario"]

plt.hist(salarios_ativos, bins=10, color="steelblue")       #cria o grafico
plt.title("Distribuição de Salários")                       #adiciona um título
plt.xlabel("Salário (R$)")                                  #titulo eixo "x"
plt.ylabel("Quantidade de Funcionários")                    #titulo eixo "y"    
plt.show()        
```
