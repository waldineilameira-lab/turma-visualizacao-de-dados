```python
#iniciar o notebook com as importações necessárias
#se aparecer erro aqui, significa que a biblioteca não está instalada, e você deve instalar usando pip install nome_da_biblioteca

import pandas as pd #biblioteca para manipulação de dados
import re #biblioteca para expressões regulares
from datetime import datetime #biblioteca para manipulação de datas

# confirmarque as bibliotecas foram importadas corretamente
print("pandas versão:", pd.__version__)
print("re versão:", re.__version__)
```

```python
#pd.read_csv() é a função para ler arquivos CSV e criar um DataFrame
#CAMINHO DO ARQUIVO: é o caminho onde o arquivo está localizado no seu computador
#basta informar o nome do arquivo se ele estiver na mesma pasta do notebook, ou o caminho completo se estiver em outra pasta
#exemplo: df = pd.read_csv("dados.csv") ou df = pd.read_csv

CAMINHO = r"C:\Users\Usuário\Downloads\SCTEC\turma-visualizacao-de-dados\aula_05_05\turma-visualizacao-de-dados\alunos\lourenco_lemos\semana_04\base_rh.csv"

df = pd.read_csv(
    CAMINHO,
    sep=";", #separador de campos, no caso é ";"
    encoding="cp1252", #codificação do arquivo, no caso é "cp1252" para arquivos em português
    decimal=",", #separador decimal, no caso é "," para arquivos em português
    thousands=".", #separador de milhares, no caso é "." para arquivos em português
    )

print("DataFrame criado com sucesso!")
print(f' total de linhas: {df.shape[0]}')
print(f' total de colunas: {df.shape[1]}')
```

```python
#df.head() mostra as primeiras 5 linhas do DataFrame, útil para verificar se os dados foram carregados corretamente
#use para ter uma visão rápida dos dados e verificar se as colunas estão corretas e os dados no formato esperado
df.head()
```

```python
#df.tail() mostra as últimas 5 linhas do DataFrame, útil para verificar se os dados foram carregados corretamente e se não há linhas em branco no final do arquivo
#use para ter uma visão rápida dos dados e verificar se as colunas estão corretas e os dados no formato esperado, especialmente se o arquivo tiver muitas linhas
df.tail(10) #mostra as últimas 10 linhas do DataFrame
```

```python
#df.columns mostra os nomes das colunas do DataFrame, útil para verificar se as colunas foram carregadas corretamente e para saber quais colunas estão disponíveis para análise
#use para ter uma visão rápida dos nomes das colunas e verificar se estão corretos, especialmente se o arquivo tiver muitas colunas
#to.list() converte o índice das colunas em uma lista, facilitando a visualização e manipulação dos nomes das colunas
#df.columns.to_list()

print("Colunas do DataFrame:")
print(df.columns.to_list())
```

```python
#df.info() mostra informações sobre o DataFrame, como o número de linhas, colunas, tipos de dados e memória utilizada
#use para ter uma visão geral do DataFrame e verificar se os tipos de dados estão corretos, se há colunas com muitos valores nulos e para ter uma ideia da estrutura dos dados
#comando mais importante para iniciar a análise dos dados, pois ajuda a identificar possíveis problemas com os dados, como colunas com muitos valores nulos ou tipos de dados incorretos
df.info()
```

```python
#df.isnull() mostra um DataFrame booleano indicando onde há valores nulos (True) e onde não há (False)
#use para identificar onde estão os valores nulos no DataFrame, o que é importante para decidir como lidar com eles (excluir, preencher, etc.)
df.isnull()
```

```python
#.sum() soma os valores booleanos, ou seja, conta quantos valores nulos há em cada coluna
#use para ter uma visão rápida de quantos valores nulos há em cada coluna, o que é importante para decidir como lidar com eles (excluir, preencher, etc.)
print("Quantidade de valores nulos por coluna:")
print(df.isnull().sum())

print()

#percentual de valores nulos em cada coluna
#use para ter uma visão rápida do percentual de valores nulos em cada coluna, o que é importante para decidir como lidar com eles (excluir, preencher, etc.)
print("Percentual de valores nulos por coluna:")
print((df.isnull().sum() / len(df)) * 100)
```

```python
#df.describe() mostra estatísticas descritivas das colunas numéricas do DataFrame, como média, desvio padrão, valores mínimos e máximos, etc.
#use para ter uma visão geral das colunas numéricas do DataFrame e identificar possíveis outliers, tendências e para ter uma ideia da distribuição dos dados
df.describe().round(2) #arredonda os valores para 2 casas decimais
```

```python
#df.nunique() mostra o número de valores únicos em cada coluna, útil para identificar colunas categóricas e para ter uma ideia da diversidade dos dados
#use para identificar colunas categóricas (que geralmente têm poucos valores únicos) e para ter uma ideia da diversidade dos dados em cada coluna
print("Número de valores únicos por coluna:")
print(df.nunique())
```
