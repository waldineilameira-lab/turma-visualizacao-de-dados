```python
# Criando um programa que simula um caixa de loja #

# 1. Peça nome de 3 produtos e seus preços
# 2. Calcule o total da compra
# 3. Aplique desconto de 10% se total > R$ 100
# 4. Exiba o valor final formatado

# Enviar via Pull Request em alunos/seu_nome/exercicio_01/



# Sistema simples de cadastro de produtos #
print("===== Sistema Integrado de Caixa da Loja =====")


# Entrada de dados
arvores_nativas = input("Nome do produto 1: ")
preco_produto1 = float(input(f"Preço do {arvores_nativas}: R$ 300,00 "))

plantas_meliferas = input("Nome do produto 2: ")
preco_produto2 = float(input(f"Preço do {plantas_meliferas}: R$ 200,00 "))
arbustos_aromaticos = input("Nome do produto 3: ")
preco_produto3 = float(input(f"Preço do {arbustos_aromaticos}: R$ 150,00 "))


# Cálculos

total = preco_produto1 + preco_produto2 + preco_produto3

desconto = 0
if total > 100:
    desconto = total * 0.10

valor_final = total - desconto


# Exibição formatada
print("\n===== Resumo da Compra =====")
print(f"{arvores_nativas}: R$ {preco_produto1:.2f}")
print(f"{plantas_meliferas}: R$ {preco_produto2:.2f}")
print(f"{arbustos_aromaticos}: R$ {preco_produto3:.2f}")
print(f"\nTotal: R$ {total:.2f}")

if desconto > 0:
    print(f"Desconto (10%): -R$ {desconto:.2f}")

print(f"Valor final: R$ {valor_final:.2f}")


# Verificação de estoque
quantidade_produto1 = int(input(f"\nQuantidade em estoque de {arvores_nativas}: "))
quantidade_produto2 = int(input(f"Quantidade em estoque de {plantas_meliferas}: "))
quantidade_produto3 = int(input(f"Quantidade em estoque de {arbustos_aromaticos}: "))

print("\n===== Situação do Estoque =====")

if quantidade_produto1 > 90:
    print(f"{arvores_nativas}: {quantidade_produto1} unidade(s) disponível(is)")
else:
    print(f"{arvores_nativas}: ESTOQUE FARTO")

if quantidade_produto2 < 40:
    print(f"{plantas_meliferas}: {quantidade_produto2} unidade(s) disponível(is)")
else:
    print(f"{plantas_meliferas}: ESTOQUE CRÍTICO")

if quantidade_produto3 < 100:
    print(f"{arbustos_aromaticos}: {quantidade_produto3} unidade(s) disponível(is)")
else:
    print(f"{arbustos_aromaticos}: COMPRAR MAIS")
```
