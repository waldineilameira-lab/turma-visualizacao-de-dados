```python
# importar bibliotecas necessárias

import pandas as pd

# ler o arquivo CSV e armazenar em um DataFrame
df = pd.read_csv('base_rh.csv',
                 sep=';',
                 encoding='cp1252',
                 decimal=',')

# EXPLORAÇÃO DOS DADOS

df.shape    # número de linhas e colunas
df.info()   # informações sobre os tipos de dados e valores nulos
df.describe()  # estatísticas descritivas para colunas numéricas
df.head()    # primeiras linhas do DataFrame
df.tail()    # últimas linhas do DataFrame 
df.columns   # nomes das colunas
df.nunique()  # número de valores únicos por coluna

df['col'].value_counts()  # contagem de valores únicos em uma coluna específica
df.isnull().sum()  # contagem de valores nulos por coluna

# SELEÇÃO E FILTROS

df['Nome']  # seleção da coluna 'Nome'
df[['Nome', 'Cargo']]  # seleção de múltiplas colunas
df[df['Salário'] > 5000]  # filtro para salários maiores que 5000
df[df['Status'] == 'Ativo']  # filtro para funcionários ativos
df[(df['Status'] == 'Ativo') & (df['Idade'] > 30)]  # filtro para funcionários ativos com idade maior que 30

# DATAS

df['Data de Admissão'] = pd.to_datetime(df['Data de Admissão'], format='%d/%m/%Y', errors = 'coerce')  # converter a coluna 'Data de Admissão' para formato datetime
df['Ano'] = df['Data de Admissão'].dt.year  # extrair o ano da data de admissão
df['Mês'] = df['Data de Admissão'].dt.month  # extrair o mês da data de admissão
df['Dia'] = df['Data de Admissão'].dt.day  # extrair o dia da data de admissão
df['Trim'] = df['Data de Admissão'].dt.quarter  # extrair o trimestre da data de admissão
df['Semana'] = df['Data de Admissão'].dt.isocalendar().week  # extrair a semana do ano da data de admissão
df['Tempo'] = ((pd.Timestamp.today() - df['Data de Admissão']).dt.days / 365.25).round(1)   # calcular o tempo em anos desde a data de admissão até hoje

novos = df[df['Data de Admissão'] >= pd.Timestamp.today() - pd.DateOffset(years=1)]  # filtrar funcionários admitidos no último ano

# AGRUPAMENTOS (GROUPBY)

df.groupby('Cargo')['Salário'].mean().round(2)  # média salarial por cargo
df.groupby('Departamento')['Salário'].size().sort_values(ascending=False)  # contagem de funcionários por departamento
df.groupby('Departamento')['Salário'].agg(['mean', 'min', 'max']).round(2)  # média, mínimo e máximo salarial por departamento
df.groupby('Status')['Idade'].median()  # idade mediana por status

# EXPORTAÇÃO DO DATAFRAME

df.to_csv('base_rh_limpa.csv', index=False, sep=';', encoding='cp1252', decimal=',')  # exportar o DataFrame para um novo arquivo CSV
df.to_json('base_rh_limpa.json', orient='records', orient='records', force_ascii=False, indent=4)  # exportar o DataFrame para um arquivo JSON

with pd.ExcelWriter('base_rh_limpa.xlsx', engine='openpyxl') as w:
    df.to_excel(w, sheet_name='Funcionários', index=False)  # exportar o DataFrame para um arquivo Excel

```
