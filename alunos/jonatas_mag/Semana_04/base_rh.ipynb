```python
# MÓDULO 10 - O DTATIME BÁSICO

# strptime() - converte string para datetime
# strftime() - converte datetime para string
# %d - dia do mês (01 a 31)
# %m - mês (01 a 12)
# %Y - ano com 4 dígitos (2024)
# %H - hora (00 a 23)
# %M - minuto (00 a 59)
# %B - nome completo do mês (Janeiro, Fevereiro, etc.)

from datetime import datetime, timedelta

data_texto = '13/08/2024' # Data em formato de string
data_obj = datetime.strptime(data_texto, '%d/%m/%Y')  # Converte a string para um objeto datetime usando o formato especificado

print(f' Data original (string): {data_texto}')
print(f' Data convertida (datetime): {data_obj}')
print(f' Tipo do objeto: {type(data_obj)}')

print(' Data transformada em string novamente com sucesso')
```

```python
# MÓDULO 11 - CONVERTENDO A COLUNA INTEIRA pd.to_datetime()

print(f'Formato Anterior: {df["Data_Admissao"].dtype}')
print(f'Exemplo de valor: {df["Data_Admissao"].iloc[0]}')

print()

# Criandop uma cópia da df para não alterar a original
# Entregar uma df_limpo através de uma df.copy()
df_limpo = df.copy()

df_limpo["Data_Admissao"] = pd.to_datetime(
    df_limpo["Data_Admissao"],
    format='%d/%m/%Y', # Especifica o formato da data
    errors='coerce'    # Converte valores inválidos para NaT (Not a Time)
)

print(f'Formato Depois: {df_limpo["Data_Admissao"].dtype}')
print(f'Valor convertido: {df_limpo["Data_Admissao"].iloc[0]}')
print(f'Número de datas inválidas convertidas para NaT: {df_limpo["Data_Admissao"].isna().sum()}')
```

```python
# MÓDULO 12 - EXTRAINDO INFORMAÇÕES DA DATA - ACESSOR .dt

df_limpo['Ano_Admissao'] = df_limpo['Data_Admissao'].dt.year
df_limpo['Mes_Admissao'] = df_limpo['Data_Admissao'].dt.month
df_limpo['Trimestre'] = df_limpo['Data_Admissao'].dt.quarter

# Data atual no formato pandas
hoje_ts = pd.Timestamp.today()
df_limpo['Tempo_Casa_Anos'] = (
    (hoje_ts - df_limpo['Data_Admissao']).dt.days / 365.25
).round(1)

# Resultado
cols = ['Nome', 'Data_Admissao', 'Ano_Admissao', 'Mes_Admissao', 'Trimestre', 'Tempo_Casa_Anos']
print(df_limpo[cols].head(10))
```

```python
# MÓDULO 13 - CRIANDO MINHA PRIMEIRA FUNÇÃO

def nome_da_funcao(parametros):
    # Bloco de código que realiza uma tarefa específica
    resultado = parametros * 2  # Exemplo de operação
    return resultado  # Retorna o resultado da função

# def - define a função
# parametro - valor de entrada da função
# return - valor de saída da função
# A identação (4 espaços) é fundamental para definir o bloco de código

# Exemplo de uso da função
resultado = nome_da_funcao(5)
print(f'O resultado da função é: {resultado}')
```

```python
# Criar a função que classifica o tempo de casa em categorias
# Facilita a interação com Gráficos e Power BI

def classificar_tempo(anos):
    """
    Recebe: anos (número) -- o tempo de casa do funcionário
    Retorna: uma categoria em texto (string)
    """
    # pd.isna() - para verificar se o valor é NaN (Not a Number)
    if pd.isna(anos):
        return 'Sem data'
    elif anos < 1:
        return 'Menos de 1 ano'
    elif anos < 3:
        return 'Entre 1 a 3 anos'
    elif anos < 5:
        return 'Entre 3 a 5 anos'
    else:
        return 'Mais de 5 anos'

# Testar a função com valores indivuduais antes e aplicar o dataframe
print('Testando a função:')
print(f'    0.5 anos -> "{classificar_tempo(0.5)}"')
print(f'    2.1 anos -> "{classificar_tempo(2.1)}"')
print(f'    4.8 anos -> "{classificar_tempo(4.8)}"')
print(f'    9.0 anos -> "{classificar_tempo(8.0)}"')
```

```python
# .apply(função) - para aplicar a função em toda a coluna do DataFrame
# É como fazer um loop para cada linha, mas de forma otimizada e rápida usando o poder do pandas

# Imprime a lista de colunas presentes no df_
print(df_limpo.columns)

df_limpo['Faixa_Tempo_Casa'] = df_limpo['Tempo_Casa_Anos'].apply(classificar_tempo)

print('Distribuição por faixa de tempo de casa:')
print(df_limpo['Faixa_Tempo_Casa'].value_counts())

```

```python
# MÓDULO 14 - FILTRAR POR PERIODO DE DATAS
# pd.dataoffset(years=1) - para criar um deslocamento de 1 ano

hoje = pd.Timestamp.today() 

# Pontos de corte
um_ano_atras = hoje - pd.DateOffset(years=1)
cinco_anos_atras = hoje - pd.DateOffset(years=5)
uma_decada_atras = hoje - pd.DateOffset(years=10)

# Filçtrar por periodo
novos     = df_limpo[df_limpo['Data_Admissao'] >= um_ano_atras]
veteranos = df_limpo[df_limpo['Data_Admissao'] < cinco_anos_atras]
senior    = df_limpo[df_limpo['Data_Admissao'] < uma_decada_atras]

print(f'Admitidos no último ano     : {len(novos)} funcionários')
print(f'Veteranos (5+ anos de casa) : {len(veteranos)} funcionários')
print(f'Sênior (10+ anos de casa)   : {len(senior)} funcionários')

print()

print('Salário médio:')
print(f'    Novos     : R$ {novos["Salario"].mean():.2f}')
print(f'    Veteranos : R$ {veteranos["Salario"].mean():.2f}')
print(f'    Sênior    : R$ {senior["Salario"].mean():.2f}')
```

```python
# MÓDULO 15 - AGRUPAMENTO COM .groupby()
# .groupby() - Tabela dinamica do Excel para agrupar os dados por uma ou mais colunas e calcular estatísticas para cada grupo
# Ex.: df.groupby('Coluna a ser agrupada')['coluna a ser calculada'].função()

# FORMULA:
# .mean()             → calcula a média de cada grupo
# .round(2)           → arredonda para 2 casas
# .sort_values(...)   → ordena do maior para o menor
# ascending=False     → decrescente (False = do maior para o menor)

print('Salário médio por departamento:')
print(
    df_limpo.groupby('Departamento')['Salario']
    .mean()
    .round(2)
    .sort_values(ascending=False)
)
```

```python
# Tempo médio de casa por departamento
media_depto = (df_limpo.groupby('Departamento')['Tempo_Casa_Anos']
    .mean()
    .round(1)
    .sort_values(ascending=False)
    .rename('Média_Anos')
    .reset_index()
)

print('Tempo médio de casa por departamento:')
print(media_depto.to_string(index=False))

# .idexmax() - para encontrar o índice do valor máximo em cada grupo
maior = media_depto.set_index('Departamento')['Média_Anos'].idxmax()
manor = media_depto.set_index('Departamento')['Média_Anos'].idxmin()
print(f'\n-> Maior Retenção: {maior}')
print(f'-> Menor Retenção: {manor}')
```

```python
# Histórico completo de Admissão por Ano

# .size() - para contar o número de registros em cada grupo
# .start_index() - para acessar o valor do grupo (ano) em cada linha do resultado

por_ano = (
    df_limpo.groupby('Ano_Admissao')
    .size()
    .rename('Admissoes')
    .reset_index()
    .dropna()
)

print('Admissões por ano:')
for _, row in por_ano.iterrows():   # Itera sobre cada linha do DataFrame
    barra = '█' * int(row['Admissoes'] / 5)  # Cria uma barra proporcional ao número de admissões
    print(f"  {int(row['Ano_Admissao'])}: {barra} {row['Admissoes']:>4} {barra}")
```

```python
# MÓDULO 16 - SALVANDO O DATSET LIMPO - ENRIQUECIDO COM NOVAS COLUNAS
# Fazer o Loard do RTL - Resultado Final (Arquivo CSV p/ o Power BI)

# Encondin = 'UTF-8-sig' - para garantir que os acentos sejam lidos corretamente no Power BI
#                          UTF-8 com BOM (Byte Order Mark) para compatibilidade total com o Power BI e Excel

df_limpo.to_csv(
    'base_rh_dia01.csv',    # Nome do arquivo de saída
    index=False,            # Não inclui o índice do DataFrame no CSV
    sep=';',                # Define o ponto e vírgula como separador de colunas
    encoding='UTF-8-sig',   # Codificação para garantir compatibilidade com acentos no Power BI
)

print('Base_rh_dia01.csv exportada com sucesso!')
print(f'    Colunas finas: {df_limpo.columns.tolist()}')
print(f'    Linhas       : {len(df_limpo)}')
```

### CONVERTENDO srt EM Datatime - ANTES e DEPOIS

```python
# Antes da conversão:
print(f'Tipo ANTES : {df["Data_Admissao"].dtype}')
print(f'Valor antes: "{df["Data_Admissao"].iloc[0]}"') # .iloc[0] → acessa a primeira linha pelo índice
print()

# Trabalhamos em uma cópia para não alterar o df original
df_ok = df.copy()
# Sem isso, mudanças em df_ok afetariam o df original

# Mudando a coluna inteira
df_ok['Data_Admissao'] = pd.to_datetime(
    df_ok['Data_Admissao'],
    format='%d/%m/%Y',  # Formato da entrada
    errors='coerce'     # Datas inválidas viram NaT (Not a Time) em vez de erro
)

print(f'Tipo DEPOIS : {df_ok["Data_Admissao"].dtype}')
print(f'Valor depois: {df_ok["Data_Admissao"].iloc[0]}')
print(f'NaT gerados : {df_ok["Data_Admissao"].isna().sum()}')
```

### FAZER A ANALISE DETALHADA DA Data_Admissao - ADD COLUNAS ESPECÍFICAS

```python
# Criar colunas derivadas da Data de Admissão
df_ok['Ano_Admissao'] = df_ok['Data_Admissao'].dt.year
df_ok['Mes_Admissao'] = df_ok['Data_Admissao'].dt.month

# Calcular tempo de casa em anos
hoje_ts = pd.Timestamp.today()      # pd.Timestamp.today() → data atual no formato do pandas

# Subtração de datas = timedelta
df_ok['Tempo_Casa_Anos'] = ((hoje_ts - df_ok['Data_Admissao']).dt.days / 365.25).round(1)    # .dt.days → extrai a diferença em dias
df_ok['Faixa_Tempo_Casa'] = ((hoje_ts - df_ok['Data_Admissao']).dt.days / 365.25).round(1)   # / 365.25 → converte dias em anos
# .round(1) → arredonda para 1 casa decimal

# Resultado final para conferência
cols = ['Nome', 'Data_Admissao', 'Ano_Admissao', 'Mes_Admissao', 'Tempo_Casa_Anos', 'Faixa_Tempo_Casa']
df_ok[cols].head(6)
```

### OBTENDO A MÉDIA DE RETENÇÃO E TEMPO DE CASA POR DEPARTAMENTO 

```python
# Tempo médio de casa por departamento
media_depto = (
    df_ok
    .groupby('Departamento')['Tempo_Casa_Anos']
    .mean()
    .round(1)
    .sort_values(ascending=False)
    .reset_index()          # .reset_index() → converte o índice em coluna comum
)

maior = media_depto.set_index('Departamento')['Tempo_Casa_Anos'].idxmax()   # .idxmax() → retorna o NOME do índice com maior valor
menor = media_depto.set_index('Departamento')['Tempo_Casa_Anos'].idxmin()   # .idxmin() → retorna o NOME do índice com menor valor
print(f'→ Maior média de tempo de casa: {maior}')
print(f'→ Menor média de tempo de casa: {menor}')
```
