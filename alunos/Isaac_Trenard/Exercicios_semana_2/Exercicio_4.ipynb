Sistema de gestão escolar -- Escola os pequenos Analistas --

```python
quantos_alunos_cadastrados = int(input('Por favor insira o numero de alunos matriculados: '))

for i in range(quantos_alunos_cadastrados):
    print(f'--- Cadastrando aluno {i+1} de {quantos_alunos_cadastrados} ---')
    
    # 1. Dados relevantes do aluno
    nome_do_aluno = input('Por favor digite o nome do aluno: ')
    idade_do_aluno = int(input('Digite a sua idade: '))
    
    # Função para calcular média
    def calcular_media_ano_escolar():
        nota_primeiro_trimestre = float(input('Digite a sua nota 1º trimestre: '))
        nota_segundo_trimestre = float(input('Digite a sua nota 2º trimestre: '))
        nota_terceiro_trimestre = float(input('Digite a sua nota 3º trimestre: '))
        
        if nota_primeiro_trimestre < 0 or nota_segundo_trimestre < 0 or nota_terceiro_trimestre < 0:
            print('Nota inválida! Contate a coordenação.')
            return None
        
        media = (nota_primeiro_trimestre + nota_segundo_trimestre + nota_terceiro_trimestre) / 3
        print(f'A média das notas do aluno {nome_do_aluno} é de {media:.2f}')
        return media
    
    # Chamar a função
    media_aluno = calcular_media_ano_escolar()
    
    # Verificar se a média é None (nota inválida)
    if media_aluno is None:
        print("Aluno ignorado devido a nota inválida.")
        continue  # Pula para o próximo aluno
    
    # Análise escolar
    if media_aluno >= 7:
        print('Aprovado!')
    elif 5 <= media_aluno < 7:
        print('Recuperação')
    elif media_aluno < 5:
        print('Reprovado')
    
    # Destaque Jovem
    if idade_do_aluno <= 15 and media_aluno >= 7:
        print('Destaque Jovem')
        print(True)
```
