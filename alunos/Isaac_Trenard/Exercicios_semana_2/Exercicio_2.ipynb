# Exercício 2 - Calculadora de Folha de Pagamento

# Crie um programa que simule o cálculo do salário líquido de um funcionário.

```python
## Etapas:
### Passo 1: Criar variáveis

nome_Do_Funcionario = input('Por favor, digite seu nome completo: ' )
salario_base = float(input('Por favor, digite seu salário exato com decimais: '))
quantidade_De_Horas_Extras = int(input('Por favor, informe a quantidade de horas extras trabalhadas: '))

print(f'Muito obrigado por informar seus dados, {nome_Do_Funcionario}!')

### Passo 2: Criar calculadora

horas_de_trabalho_mes = 220
porcentual_hora_extra = 1.5
valor_hora_extra = (salario_base / 220) * 1.5
hora_extra_ganha = quantidade_De_Horas_Extras * valor_hora_extra
print (f'Você  trabalhou, {quantidade_De_Horas_Extras} e o valor ganho foi de, {hora_extra_ganha:.2f}')
salario_base_com_hora_extra = salario_base + hora_extra_ganha
print(f'Parabéns! Você vai receber R$ {salario_base_com_hora_extra:.2f} no seu salário por ser uma pessoa esforçada.')

### Passo 3: Calcular desconto do INSS (previdência)

calculo_inss_Desconto = (100 - 11) /100 #0.89 porncetual

Salario_com_desconto_inss = salario_base_com_hora_extra * calculo_inss_Desconto

print(f'Porém, você precisa pagar o INSS. Seu salário com desconto ficará em R$ {Salario_com_desconto_inss:.2f}')

### Passo 4: Verificar se o funcionário tem direito a bônus
## O funcionário ganha um bônus se:
## - Salário final for **maior que R$ 3.000,00** 
## - **E** tiver **mais de 10 horas extras** trabalhadas

if salario_base_com_hora_extra >= 3000.00:
    print(True)
    print('Parabéns! Você ganhou um dia de folga!')

else:
    print('Você pode trabalhar meio período na sexta-feira.')
    print(False)

if hora_extra_ganha >= 10 and salario_base_com_hora_extra >= 2000.00:
    print ('Você é um excelente programador!')
```
