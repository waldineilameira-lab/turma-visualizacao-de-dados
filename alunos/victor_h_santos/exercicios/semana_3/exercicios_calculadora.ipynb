# Exercício 1: Calculadora
Crie um programa que pede dois números ao usuário e exibe:

* Soma
* Subtração
* Multiplicação
* Divisão


```python
resultado = 0
while True:
    try:
        valor1 = float(input("Digite o primeiro valor: "))
        valor2 = float(input("Digite o segundo valor: "))
        operacao = input("Digite a operação (+, -, *, /): ")
        if operacao == "+":
            resultado = float(valor1) + float(valor2)
        elif operacao == "-":
            resultado = float(valor1) - float(valor2)
        elif operacao == "*":
            resultado = float(valor1) * float(valor2)
        elif operacao == "/":
            resultado = float(valor1) / float(valor2)
        else:
            print("Operação inválida. Tente novamente.")
            continue
    except ValueError:
        print("Digite apenas numeros. Tente novamente.")
            
    except ZeroDivisionError:
        print("Não é possível dividir por zero. Tente novamente.")
    else:
        print("Resultado:", resultado)
        break
```

# Exercício 2: Par ou Ímpar
Peça um número ao usuário e informe se é par ou ímpar.

Dica: Use o operador % (módulo)
Se numero % 2 == 0 é par!

```python

while True:
    try:
        valor1 = int(input("Digite o primeiro valor: "))
        if valor1 % 2 != 0:
            print("Seu numero é ímpar.")
            break
        else:
            print("Seu numero é par.")
            break
    except ValueError:
        print("Digite apenas numeros. Tente novamente.")
```

# Exercício 3: Média de Notas
Peça 3 notas ao usuário, calcule a média e informe:
* Média >= 7: "Aprovado"
* Média < 7: "Reprovado"

```python
while True:
    try:
        valor1 = int(input("Digite o valor para saber se é impar ou par: "))
        if valor1 % 2 != 0:
            print("Seu numero é ímpar.")
            break
        else:
            print("Seu numero é par.")
            break
    except ValueError:
        print("Digite apenas numeros. Tente novamente.")
```

# Exercício 4: Tabuada
Peça um número e exiba a tabuada dele de 1 a 10.
Use um loop for com range(1, 11)

```python
while True:
    try:
        valor1 = int(input("Digite o valor para saber a tabuada: "))
        for i in range(1, 11):
            print(f"{valor1} x {i} = {valor1 * i}")
        break
    except ValueError:
        print("Digite apenas numeros. Tente novamente.")
```
