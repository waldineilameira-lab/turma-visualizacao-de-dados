# EXERCÍCIO 3 — Sistema de Classificação de IMC

### Enunciado:

Crie um programa que calcule o IMC de uma pessoa e classifique o resultado:

1. Peça: nome, peso (`float`) e altura (`float`)
2. Calcule o IMC: `peso / (altura ** 2)`
3. Classifique usando `elif`:
    * Abaixo de 18.5: Abaixo do peso
    * 18.5 a 24.9: Peso normal
    * 25.0 a 29.9: Sobrepeso
    * 30.0 a 34.9: Obesidade grau 1
    * 35.0 a 39.9: Obesidade grau 2
    * 40 ou mais: Obesidade grau 3
4. Verifique com `and` se o IMC é normal E se a pessoa tem mais de 60 anos — se sim, exiba uma mensagem especial de atenção à saúde
5. Exiba se o IMC está acima de 25 (`True`/`False` usando operador relacional)
