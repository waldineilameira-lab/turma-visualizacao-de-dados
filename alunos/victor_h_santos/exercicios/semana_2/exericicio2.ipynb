# EXERCÍCIO 2 — Calculadora de Folha de Pagamento

### Enunciado:

Crie um programa que simule o cálculo de salário líquido de um funcionário:

1. Peça: nome do funcionário, salário bruto (`float`) e quantidade de horas extras trabalhadas (`int`)
2. Calcule: valor de cada hora extra = salário bruto dividido por 220 horas, multiplicado por 1.5
3. Calcule o total de horas extras e some ao salário bruto
4. Aplique desconto de INSS de 11% sobre o salário bruto
5. Verifique (com operadores relacionais) se o salário final é maior que R$ 3.000,00 e exiba `True` ou `False`
6. Verifique se o funcionário tem mais de 10 horas extras E salário acima de 2.000 (use `and`)
7. Exiba um relatório completo com todos os valores e resultados das comparações
