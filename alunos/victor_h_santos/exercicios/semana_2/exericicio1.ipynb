# EXERCÍCIO 1 — Ficha Cadastral

Agora é com vocês. Leiam com atenção antes de começar a digitar.

### Enunciado:

Crie um programa no Google Colab que faça o seguinte:

1. Peça ao usuário para digitar: nome, cidade, profissão, idade e salário
2. Guarde cada informação em uma variável com nome adequado
3. Use `type()` para verificar e exibir o tipo de cada variável
4. Mostre uma ficha cadastral completa usando f-string, com todas as informações organizadas numa mensagem clara
5. Ao final, exiba quanto o usuário vai ganhar daqui a 3 anos se o salário aumentar R$ 300,00 por ano

**Dica:** Lembre de converter idade para `int` e salário para `float`.
