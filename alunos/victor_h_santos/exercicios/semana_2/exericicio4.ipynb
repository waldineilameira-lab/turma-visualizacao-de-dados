# EXERCÍCIO FINAL DA SEMANA 02 — Sistema Completo de Gestão Escolar
Este é o exercício final do curso. Ele integra absolutamente tudo que vocês aprenderam nos três dias.

## Enunciado:
Crie um sistema de gestão escolar que faça o seguinte:

### Parte 1 — Cadastro:
* Pergunte quantos alunos serão cadastrados
* Para cada aluno, colete: nome, idade (int) e três notas (float cada uma)

### Parte 2 — Processamento:
Para cada aluno, calcule a média das três notas

Classifique com if/elif:

* Aprovado (média >= 7)
* Recuperação (5 <= média < 7)
* Reprovado (média < 5)

Verifique se o aluno tem menos de 15 anos AND está aprovado — se sim, exiba "Destaque jovem!"

Use continue para pular o processamento se qualquer nota for negativa (nota inválida)

### Parte 3 — Relatório:
Após processar todos os alunos, exiba:

* Total de alunos cadastrados
* Quantidade de aprovados, em recuperação e reprovados
* Maior média da turma e o nome do aluno
* Média geral da turma
* Se a média geral for menor que 6, exiba alerta de turma em risco

Dica: Use acumuladores para cada contador e para a soma das médias.
