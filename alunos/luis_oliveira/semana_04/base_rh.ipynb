```python
# Cria uma tabela de True/False indicando a presença de valores nulos em cada célula do DataFrame usando o método isnull() do pandas
# .sum() para obter a soma total de nulos por coluna, ou seja, quantos valores nulos existem em cada coluna do DataFrame

print('Quantidade de valores nulos por coluna:  ')
print(df.isnull().sum())    



# Percentual de nulo 
# len(df) para obter o número total de linhas do DataFrame, e calcular o percentual de valores nulos em cada coluna dividindo a quantidade de nulos pela quantidade total de linhas e multiplicando por 100

print('Percentual de valores nulos por coluna:  ')
print((df.isnull().sum() / len(df)) * 100)
```
