```python
import pandas as pd

# ----- LEITURA DO ARQUIVO CSV -----
df = pd.read_csv("base_rh.csv", sep=";", encoding="cp1252", decimal=",") 

# ----- EXPLORAÇÃO DOS DADOS -----
df.shape    #(1000, 10) - 1000 linhas e 10 colunas
df.head(5)  # Exibe as 5 primeiras linhas
df.tail(5)  # Exibe as 5 últimas linhas
df.info()   # Exibe informações sobre o DF (tipos de dados, valores) 
df.describe()   # Exibe estatísticas descritivas para as colunas numéricas do DF
df.nunique()    # Exibe o número de valores únicos em cada coluna do DF

df['col'].value_counts()   # Exibe a contagem de valores únicos na coluna 'col'
df.isnull().sum()   # Exibe a contagem de valores nulos em cada coluna do DF

# ----- SELEÇÃO E FILTROS -----
df['Nome']  # Seleciona a coluna 'Nome' do DF
df[['Nome', 'Idade']]   # Seleciona as colunas 'Nome' e 'Idade" do DF
df[df['Status'] == 'Ativo'] # Filtro simples para selecionar apenas as linhas onde a coluna 'Status'...
df[(df['Status'] == 'Ativo') & (df['Idade'] > 30)]   # Filtro composto para selecionar apenas as linhas...

# ----- DATAS -----
df['Data de Admissão'] = pd.to_datetime(df['Data de Admissão'], format='%d/%m/%Y', errors='coerce')
df["Ano"] = df['Data de Admissão'].dt.year
df["Mês"] = df['Data de Admissão'].dt.month
df["Trim"] = df['Data de Admissão'].dt.quarter
df["Tempo"] = ((pd.Timestamp.today() - df["Data de Admissão"]).dt.days / 365.25).round(1)

# ----- GROUPBY -----
df.groupby('Departamento')['Salário'].mean().round(2)
df.groupby('Departamento')['Salário'].size().sort_values(ascending=False)

# ----- EXPORTAÇÃO DO DATAFRAME -----
df.to_csv("saida.csv", index=False, sep=";", encoding='utf-8')
df.to_json("saida.json", orient="records", force_ascii=False, indent=4)

#with pd.ExcelWriter("saida.xlsx", engine="openpyxl") as 
```
