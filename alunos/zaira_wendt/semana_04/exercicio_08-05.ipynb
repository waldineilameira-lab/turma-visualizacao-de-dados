```python
# ── Resolva o exercício aqui ──────────────────────────────────────────────
# Passo 1: carregue o CSV

import pandas as pd 
import re   
from datetime import datetime 

df = pd.read_csv(
    'base_rh.csv',
    sep=';',
    encoding='cp1252',
    decimal=','
)





# Passo 6: cole o output do git log aqui como comentário
# git log:
```

```python
# Passo 2: diagnóstico 
df.info() # Resumo completo do DataFrame em uma só chamada - Exibe informações do DataFrame, como tipos de dados e valores.

print("Quantidade de nulos por coluna: ")
print(df.isnull().sum())

print()
```

```python
# Passo 3: converter datas e criar colunas derivadas 
df['Data_Admissao'] = pd.to_datetime(df['Data_Admissao'], format='%d/%m/%Y', errors='coerce')
df["Ano"] = df["Data_Admissao"].dt.year
df["Mês"] = df["Data_Admissao"].dt.month
df["trim"] = df["Data_Admissao"].dt.quarter
df["Tempo"] = ((pd.Timestamp.today() - df["Data_Admissao"]).dt.days / 365.25).round(1)
```

```python
# Passo 4: groupby — maior e menor média de tempo de casa 

df.groupby('Departamento')['Salario'].mean().round(2)   
df.groupby('Departamento')['Salario'].size().sort_values(ascending=False) 
```

```python
# Passo 5: exportar para Excel 

df.to_excel('analise_rh.xlsx', index=False)
```
