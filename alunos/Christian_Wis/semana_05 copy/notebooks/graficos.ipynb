```python
import pandas as pd
import matplotlib.pyplot as plt


# Estilo do gráfico (preview deles na documentação do Matplotlib)

# plt.style.use("dark_background")
# plt.style.use("ggplot")
# plt.style.use("bmh")
# print(plt.style.available)

# plt.style.use("ggplot")
# plt.style.use("seaborn-v0_8")
# plt.style.use("dark_background")
# plt.style.use("bmh")
# plt.style.use("fivethirtyeight")

URL = (
    "https://raw.githubusercontent.com/"
    "cfneves/turma-visualizacao-de-dados/"
    "master/aulas/semana_04/bases/base_rh.xlsx"
)

df = pd.read_excel(URL)

df["Data_Admissao"] = pd.to_datetime(
    df["Data_Admissao"],
    format="%d/%m/%Y",
    errors="coerce"
)

# Tamanho da figura
plt.figure(figsize=(10,5))

# Histograma
plt.hist(
    df["Salario"],
    bins=10,
    color="steelblue",
    edgecolor="white"
)

# Títulos
plt.title("Distribuição de Salários", fontsize=14)
plt.xlabel("Salário (R$)")
plt.ylabel("Qtde de Funcionários")

plt.show()


```

# grafico de barras verticais

```python
plt.hist(df['Salario'], bins=25, color='#0066ff', edgecolor='#000000')
plt.title('1. Histograma: Distribuição de Salários')
plt.xlabel('Salário (R$)')
plt.ylabel('Frequência')
plt.tight_layout()
plt.show()
```

# grafico de barras horizontais

```python
media_depto = df.groupby('Departamento')['Salario'].mean().sort_values()
plt.barh(media_depto.index, media_depto.values, color='#0037ff')
plt.title('2. Barras Horizontais: Média Salarial por Depto')
plt.xlabel('Salário Médio (R$)')
plt.tight_layout()
plt.show()
```

# grafico de pizza

```python
cores = ['#db6464', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']

estado_civil = df['Estado_Civil'].value_counts()
plt.pie(estado_civil.values, labels=estado_civil.index, autopct="%1.1f%%", startangle=90, colors=cores)
plt.title('3. Pizza: Distribuição por Estado Civil')
plt.tight_layout()
plt.show()
```

# grafico de dispersão

```python
plt.scatter(df['Idade'], df['Salario'], alpha=0.6, color='#6864db')
plt.title('4. Dispersão: Idade vs Salário')
plt.xlabel('Idade (anos)')
plt.ylabel('Salário (R$)')
plt.tight_layout()
plt.show()
```

# gráfico de linha

```python
admissoes_por_ano = df['Data_Admissao'].dt.year.value_counts().sort_index()
plt.plot(admissoes_por_ano.index, admissoes_por_ano.values, marker='o', linestyle='-', color='#0008ff')
plt.title('5. Linha: Contratações por Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Contratações')
plt.xticks(admissoes_por_ano.index, rotation=45)
plt.tight_layout()
plt.show()
```

# grafico de distribuição

```python
generos_list = df['Genero'].unique()
dados_boxplot = [df[df['Genero'] == g]['Salario'] for g in generos_list]

plt.boxplot(dados_boxplot, labels=generos_list, patch_artist=True, boxprops=dict(facecolor='#00b3ff'))
plt.title('6. Boxplot: Salário por Gênero')
plt.ylabel('Salário (R$)')
plt.tight_layout()
plt.show()
```
