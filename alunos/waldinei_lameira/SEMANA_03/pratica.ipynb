#### Exercício 1: Calculadora
Crie um programa que pede dois números ao usuário e exibe:
 Soma
 Subtração
 Multiplicação
 Divisão

```python
# Entrada de dados

print("\n------ CALCULADORA SIMPLES -------")

num1 = float(input("Informe o primeiro número: "))
num2 = float(input("Informe o segundo número: "))

# Operações

som = num1 + num2
sub = num1 - num2
multp = num1 * num2

# Restrigindo a divisão por zero

if num2 != 0:
    div = num1 / num2
else:
    div = "Não definida a divisão por 0"

# Resultados das operações

print("\n ====== RESULTADOS ======")
print(f"Soma         {num1} + {num2} = {som}")
print(f"Subtração         {num1} - {num2} = {sub}")
print(f"Multiplicação         {num1} * {num2} = {multp}")
print(f"Divisão         {num1} / {num2} = {div}")
```

#### Exercício 2: Par ou Ímpar 

Peça um número ao usuário e informe se é par ou ímpar.

Dica: Use o operador % (módulo)
Se numero % 2 == 0 é par!

```python
# Entrada dos dados

numero = int(input("Informe o seu número: "))

# Verificação se o número é par ou ímpar

if numero %2 == 0:
    print(f"O número {numero} é par")
else:
    print(f"O número {numero} é ímpar")
```

#### Exercício 3: Média de Notas
Peça 3 notas ao usuário, calcule a média e informe:

Média >= 7: "Aprovado"

Média < 7: "Reprovado"

```python
# Entrada dos dados

n1 = float(input("Informe a primeira nota: "))
n2 = float(input("Informe a segunda nota: "))
n3 = float(input("Infome a terceira nota: "))

# Cálculo da média

x = (n1 + n2 + n3) / 3

# Condição e resultado

print("\n =========== RESULTADO ===========")

if x >= 7:
    print(f"Primeira nota {n1} ")
    print(f"Segunda nota {n2} ")
    print(f"Terceira nota {n3} ")
    print(f"A sua média foi {x:.2f}, parabésn você foi APROVADO 🟩")
else:
    print(f"A sua média foi {x:.2f}, infelizmente você foi REPROVADO 🟥")
```

#### Exercício 4: Tabuada

Peça um número e exiba a tabuada dele de 1 a 10.

Use um loop for com range(1, 11)

```python
# Entrada dos dados

numero = int(input("Informe o numero para exibirmos a tabuada: "))


# Exibição da tabuada com o loop for

print(f"\n Tabuada do número {numero}:")
print("=" * 20)

for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i:2} = {resultado}")
  
```
