```python
import pandas as pd
import re # expressões regulates (para limpar texto)
from datetime import datetime # bib para trabalhar com datas

# testando as importações
print('Pandas versão:', pd.__version__)
print('Bibliotecas importadas')
```

```python
# importando o arquivo CSV - locao do arquivo CSV
CAMINHO = R"C:\turma-visualizacao-de-dados\alunos\sergio_leite\semana_04\base_rh.csv"

# pd.read_csv vai ler e add os dados de leitura a variavel df que será usada para acessar o arquivo e gerar dados

df = pd.read_csv(
    CAMINHO,
    sep=';', # padrão de ; para separar colunas e não virgula
    encoding='cp1252', # encoding do windows para leitura de acentos
    decimal=',' # add padrão de virgula para separar decimais
)

print('Dados carregados corretamente')
print(f'Total de linhas: {df.shape[0]}') # df.shape[0] se refere ao numero de linhas
print(f'Total de colunas: {df.shape[1]}') # df.shape[1] se refere ao numero de colunas
```

```python
print('Seguem as 5 primeiras linhas') # ajuda a verificar se as linhas iniciais foram lidas corretamente
print(df.head(5)) # chama a var df e solicita os 5 primeiras linhas do inicio

print('Seguem as últimas 5 linhas') # ajuda a verificar se as linhas finais foram lidas corretamente
print(df.tail(5)) # solicita ao df as últimas 5 linhas
```

```python
# df.columns mostra as colunas do dataset
# tolist transforma o index de colunas em uma lista e ajuda a acessar para fazer modificações em nomes, por exemplo.

print(df.columns.tolist()) # mostra as colunas do dataset
```

```python
matriz_dados = df.values.tolist() # cria a matriz de dados sendo uma lista python com todos os dados do dataframe

print(matriz_dados[9])
```

```python
# df.info() faz o resumo completo do DF em um comando
# mostra o nome de cada coluna, quantidade de valores nao nulos, tipos de dados (dtype), uso de memoria 
# é um comando muito útil para um primeiro diagnóstico da base de dados
df.info() 
```

```python
# para saber a quantidade de dados nulos no DF
# o df.isnull cria uma tabela true / false onde ons true são os nulos
# .sum() vai somar os true de cada coluna (true 1 e false 0)

print('Quantidade de dados nulos por coluna no data frame')
print(df.isnull().sum())
print() # pular uma linha

# calculando o percentual de nulos
# len(df) vai pegar a quandidade de linhas do DF
# calcular o percentual de nulos é muito útil em grandes bases para entender a proporção de nulos

print('O percentual de nulos no DF é: ')
print((df.isnull().sum() / len(df) * 100).round(2)) 
```

```python
# O describe() é muito usado para ver a estatistica descritiva das colunas numéricas 
# ajuda a confirmar que dados estão no formato esperado, sem valores absurdos e sem erros de leituras ou outros erros.
# count - quantos valores não nulos existem
# mean - média
# std - Desvio padrão - o quanto os valores variam em torno da média
# min - Menor valor
# 25% - 25% dos registros estão abaixo desse valor
# 50% - mediana. 50% dos registros abaixo de 50%
# 75% - 75% dos registros estão abaixo desse valor
# max - Maior valor

df.describe().round(2) # o round(2) limita o valor a 2 casas decimais
```

```python
# Explorando colunas
# # df.nunique() vai mostrar valores únicos por coluna
# Ajuda a identificar quais são categorias (poucas opções) e quais sao identificadores (muitas opções diferentes)
# nunique() - conta quantos valores diferentes existem 
# value_counts() mostra a distribuição - quantas vezes cada valor aparece
print('Valores únicos por coluna')
print(df.nunique())

```

```python
# Explorando colunas
# df['Salario'].value_counts() #  conta quantas vezes cada valor apaece na coluna escolhida

for col in ['Departamento', 'Cargo', 'Genero', 'Status', 'Estado_Civil']:
    print(f'\n --- {col} ---')
    print(df[col].value_counts())
```

```python
# Selecionando colunas e filtrando linhas
# df['Nome'] - seleciona a coluna
# df['Nome', 'Salario'] - Seleciona váriadas colunas
# df[df['Status'] == 'Ativo'] - Filtra de linhas

df['Departamento'].head(3) # lista os 3 primeiros departamentos
df['Departamento'].value_counts() # conta quantas vezes o valor aparece repetido na coluna
```

```python
# Selecionando multiplas colunas

df[['Nome', 'Cargo', 'Departamento']].value_counts()

df[['Nome', 'Cargo', 'Departamento']].head(10)
```

```python
# Filtro simples: Seleciona somente as linhas onde o status == 'Ativo'
# df['Status'] == 'Ativo' - cria uma serie true / false
# df[True / False] = mantém so as linhas true

ativos = df[df['Status'] == 'Ativo']
print(f'Total de funcionários: {len(df)}')
print(f'Funcionários ativos: {len(ativos)}')
print(f'Funcionários inativos: {len(df) - len(ativos)}')
```
