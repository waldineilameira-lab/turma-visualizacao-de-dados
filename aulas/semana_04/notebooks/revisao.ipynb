```python
import pandas as pd

#---- LEITURA DO ARQUIVO CSV ----
df = pd.read_csv("base_rh.csv", sep=";", encoding="cp1252", decimal=",")

# -- EXPLORAÇÃO DOS DADOS --
df.shape        #(1000, 10) - 1000 linhas e 10 colunas
df.head(5)      #exibe as 5 primeiras linhas do DataFrame
df.tail(5)      #exibe as 5 últimas linhas do DataFrame
df.info()       #exibe informações sobre o DataFrame, como tipos de dados e valores
df.describe()   #exibe estatísticas descritivas para as colunas numéricas do DataFrame
df.nunique()     #exibe o número de valores únicos em cada coluna do DataFrame

df['col'].value_counts()   #exibe a contagem de valores únicos na coluna 'col'
df.isnull().sum()          #exibe a contagem de valores nulos em cada coluna do DataFrame


# --- SELEÇÃO E FILTROS ---
df['Nome']     #seleciona a coluna 'Nome' do DataFrame
df[['Nome', 'Idade']]   #seleciona as colunas 'Nome' e 'Idade' do DataFrame
df[df['Status'] == 'Ativo']   # filtro simples para selecionar apenas as linhas onde a coluna 'Status' é igual a 'Ativo'
df[(df['Status'] == 'Ativo') & (df['Idade'] > 30)]   #filtro composto para selecionar apenas as linhas onde a coluna 'Status' é igual a 'Ativo' e a coluna 'Idade' é maior que 30

# --- DATAS ---
df['Data de Admissão'] = pd.to_datetime(df['Data de Admissão'], format='%d/%m/%Y', errors='coerce')
df["Ano"] = df["Data de Admissão"].dt.year     
df["Mes"] = df["Data de Admissão"].dt.month    
df["Trim"] = df["Data de Admissão"].dt.quarter    
df["Tempo"] = ((pd.Timestamp.today() - df["Data de Admissão"]).dt.days / 365.25).round(1) 

novos = df[df["Data de Admissão"] >= pd.Timestamp.today() - pd.DateOffset(years=1)] #seleciona apenas as linhas onde a coluna 'Data de Admissão' é maior ou igual a um ano atrás

# --- GROUPBY ---
df.groupby('Departamento')['Salário'].mean().round(2)  #agrupa os dados por 'Departamento' e calcula a média do 'Salário' para cada grupo
df.groupby('Departamento')['Salário'].size().sort_values(ascending=False)  #agrupa os dados por 'Departamento' e conta o número de ocorrências para cada grupo, ordenando em ordem decrescente

# --- EXPORTAÇÃO DO DATAFRAME ---
df.to_csv("saida.csv", index=False, sep=";", encoding="utf-8")
df.to_json("saida.json", orient="records", force_ascii=False, indent=4) 

with pd.ExcelWriter("saida.xlsx", engine="openpyxl") as w:
    df.to_excel(w, sheet_name="Dados", index=False)



```
